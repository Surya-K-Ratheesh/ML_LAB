```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv("Ice_cream_selling_data.csv")
print(data.head())
print(data.describe())
```

```python
# Use the correct column names
X = data['Temperature (°C)'].values.reshape(-1, 1)
Y = data['Ice Cream Sales (units)'].values
```

```python
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, Y)
```

```python
Y_pred = model.predict(X_poly)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, label='Data')
X_sort = np.sort(X, axis=0)
X_sort_poly = poly.transform(X_sort)
Y_sort = model.predict(X_sort_poly)
plt.plot(X_sort, Y_sort, color='red', label='Polynomial Regression')
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales (units)')
plt.title('Polynomial Regression: Ice Cream Sales vs Temperature')
plt.legend()
plt.show()
```
